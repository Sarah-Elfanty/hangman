```python
import random
# Write your code here

def play():

    tries = 0
    winner = 0
    words_list = ['python', 'java', 'kotlin', 'javascript']




    word = "javascript"


    guess = "-" * len(word)
    #print("\n"+guess)
    orig_len = len(word)
    letters = set()
    while tries < 8:
        if not winner:
            print("\n"+guess)
        print("Input a letter: > ")
        in_char = input()
        if len(list(in_char)) > 1:
            print("You should input a single letter")
        else:
            if not in_char.islower():
                print("It is not an ASCII lowercase letter")
            else:
                if in_char in letters:
                    print("You already typed this letter")
                else:
                    letters.add(in_char)
                    substring = word
                    indeces = []
                    while substring.find(in_char)!= -1:
                        index = substring.find(in_char)
                        indeces.append(index + (orig_len - len(substring)))
                        substring = substring[(index + 1):]
                    if len(indeces) != 0:
                        guess = list(guess)
                        for index in indeces:       
                            guess[index] = in_char
                        guess = ''.join(guess)
                        if '-' not in guess:
                            winner = 1
                            break
                    else:
                        tries += 1
                        print("No such letter in the word")

    if winner == 1:
        print("""You guessed the word!
    You survived!""")
    else:
        print("You are hanged!")

```
